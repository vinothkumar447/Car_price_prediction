```python
def get_brand_name(car_name):
    # Split the car name by spaces and take the first word (brand)
    car_name = car_name.split(' ')[0]
    
    # Return the brand name with any leading or trailing whitespace removed
    return car_name.strip()
```

```python
def clean_data(value):
    # Split the input value by spaces and take the first word
    value = value.split(' ')[0]
    
    # Remove any leading or trailing whitespace from the value
    value = value.strip()
    
    # If the value is empty after stripping, set it to 0
    if value == '':
        value = 0
    
    # Return the value as a float
    return float(value)
```

##### Separating the features & Target

```python
# input_data contains all columns except the target variable 'selling_price'
input_data = car_price.drop(columns=['selling_price'])

# output_data contains only the target variable 'selling_price'
output_data =car_price['selling_price']
```

##### Splitting the data to training data & Test data

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2)
```

##### Model Training

```python
# from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso

# Create an instance of the DecisionTreeClassifier
model = Lasso()
model.fit(x_train,y_train)
```

```python
# Make predictions 
y_pred_class = model.predict(x_test)  # For classifiers
```

##### Model Evaluation

##### Accuracy Score

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Make predictions
predict = model.predict(x_test)

# Calculate error metrics
mae = mean_absolute_error(y_test, predict)
mse = mean_squared_error(y_test, predict)
rmse = mean_squared_error(y_test, predict, squared=False)  # RMSE
r2 = r2_score(y_test, predict)

# Print results
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R Score: {r2:.2f}")  
```

##### Saving the trained model

```python
import pickle

# Save the trained model to a file
with open('car_price_model.pkl', 'wb') as file:
    pickle.dump(model, file)

print("Model saved successfully!")
```
